```python
from arcgis import *
from IPython.display import display
```

```python
gis = GIS('https://citygov.maps.arcgis.com','phammons_citygov')
```

```python
items = gis.content.search('Philadelphia', max_items=10)
```

```python
for item in items:
    display(item)
```

```python
philly = gis.map('Philadelphia')
philly
```
